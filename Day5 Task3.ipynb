# Task 3: Normalize and Transform Data

## Step 1: Prepare the dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from scipy.stats import boxcox

# Generate example data (replace with your actual dataset)
np.random.seed(42)
data = np.random.gamma(shape=2, scale=2, size=(1000, 2))
```

## Step 2: Apply Min-Max scaling and Standardization to a dataset.

```python
# Apply Min-Max scaling
scaler_minmax = MinMaxScaler()
data_minmax = scaler_minmax.fit_transform(data)

# Apply Standardization
scaler_std = StandardScaler()
data_std = scaler_std.fit_transform(data)
```

## Step 3: Perform log, square root, and Box-Cox transformations on skewed data columns

```python
# Apply Log transformation
data_log = np.log(data + 1)  # Adding 1 to avoid log(0) errors

# Apply Square Root transformation
data_sqrt = np.sqrt(data)

# Apply Box-Cox transformation
data_boxcox, _ = boxcox(data.flatten())
```

## Step 4: Visualize the effects of these transformations on data distribution

```python
# Visualization
fig, axs = plt.subplots(3, 2, figsize=(12, 12))

# Original data
axs[0, 0].hist(data[:, 0], bins=20, color='skyblue', edgecolor='black')
axs[0, 0].set_title('Original Data')

# Min-Max scaled data
axs[0, 1].hist(data_minmax[:, 0], bins=20, color='skyblue', edgecolor='black')
axs[0, 1].set_title('Min-Max Scaled')

# Standardized data
axs[1, 0].hist(data_std[:, 0], bins=20, color='skyblue', edgecolor='black')
axs[1, 0].set_title('Standardized')

# Log transformed data
axs[1, 1].hist(data_log[:, 0], bins=20, color='skyblue', edgecolor='black')
axs[1, 1].set_title('Log Transformed')

# Square root transformed data
axs[2, 0].hist(data_sqrt[:, 0], bins=20, color='skyblue', edgecolor='black')
axs[2, 0].set_title('Square Root Transformed')

# Box-Cox transformed data
axs[2, 1].hist(data_boxcox, bins=20, color='skyblue', edgecolor='black')
axs[2, 1].set_title('Box-Cox Transformed')

plt.tight_layout()
plt.show()
```
